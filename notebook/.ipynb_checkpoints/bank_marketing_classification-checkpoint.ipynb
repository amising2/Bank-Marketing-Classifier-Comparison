```python
# 1. Exploratory Data Analysis (EDA)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
df = pd.read_csv('../data/bank-additional-full.csv', sep=';')
df.head()
```

```python
# Target distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='y')
plt.title('Target Variable Distribution')
plt.xlabel('Subscribed to Term Deposit')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(6,4))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Categorical feature visualization (e.g., job)
plt.figure(figsize=(10,5))
sns.countplot(data=df, y='job', order=df['job'].value_counts().index)
plt.title('Job Distribution')
plt.xlabel('Count')
plt.ylabel('Job')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap (numeric only)
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
df.drop(columns=['duration'], inplace=True)
df['y'] = df['y'].map({'no': 0, 'yes': 1})
df = pd.get_dummies(df, drop_first=True)
```

```python
# 3. Train/Test Split
X = df.drop('y', axis=1)
y = df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
#4. Define Models and Parameters
models = {
    "Logistic Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(C=1, max_iter=1000))
    ]),
    "KNN": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", KNeighborsClassifier(n_neighbors=5))
    ]),
    "Decision Tree": Pipeline([
        ("scaler", StandardScaler()),  # optional for tree
        ("clf", DecisionTreeClassifier(max_depth=5))
    ]),
    "SVM (RBF Kernel)": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", SVC(probability=True, C=1, kernel='rbf'))
    ])
}

# Plot ROC curves
plt.figure(figsize=(8, 6))

for name, model in models.items():
    model.fit(X_train, y_train)
    y_scores = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_scores)
    plt.plot(fpr, tpr, label=name)

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot ROC curves with AUC values in legend
plt.figure(figsize=(8, 6))

for name, model in models.items():
    model.fit(X_train, y_train)
    y_scores = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_scores)
    auc = roc_auc_score(y_test, y_scores)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.3f})")

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve with AUC")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Store evaluation reports
evaluation_reports = {}

# Compute and collect confusion matrices and F1-scores
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    f1_score = report['1']['f1-score']
    evaluation_reports[name] = {
        "confusion_matrix": cm,
        "f1_score": f1_score,
        "report": report
    }

import pandas as pd
from IPython.display import display

# Display F1 scores summary
f1_df = pd.DataFrame({
    "Model": list(evaluation_reports.keys()),
    "F1 Score (class=1)": [round(v["f1_score"], 3) for v in evaluation_reports.values()]
}).sort_values(by="F1 Score (class=1)", ascending=False)

f1_df
```

```python
# Plot confusion matrix heatmaps for each model
for name, result in evaluation_reports.items():
    cm = result["confusion_matrix"]
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Predicted: No', 'Predicted: Yes'],
                yticklabels=['Actual: No', 'Actual: Yes'])
    plt.title(f'Confusion Matrix: {name}')
    plt.xlabel('Prediction')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()
```
